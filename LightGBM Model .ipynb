```python
tr_X = train.drop(['id','satisfied'],axis=1)
tr_y = train['satisfied']
```

```python
params = {    
          'boosting_type': 'gbdt',
          'objective': 'binary',
          'metric': 'auc',
          'nthread':4,
          'learning_rate':0.1,
          'num_leaves':30, 
          'max_depth': 5,   
          'subsample': 0.8, 
          'colsample_bytree': 0.8, 
    }
```

```python
import lightgbm as lgb
data_train = lgb.Dataset(tr_X, tr_y, silent=True)
```

```python
cv_results = lgb.cv(params, data_train, num_boost_round=1000, nfold=5, stratified=False, shuffle=True, metrics='auc',early_stopping_rounds=50,seed=0)
print('best n_estimators:', len(cv_results['auc-mean']))
print('best cv score:', pd.Series(cv_results['auc-mean']).max()) 
```

```python
from sklearn.model_selection import cross_validate, GridSearchCV

import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 12, 4
import matplotlib as mpl

import warnings
warnings.filterwarnings("ignore")
```

```python
param = range(0,8,1)
parameters ={'max_depth':param}
gsearch1 = GridSearchCV(estimator = lgb.LGBMClassifier(boosting_type='gbdt',objective='binary',metrics='auc',learning_rate=0.1,
                                                       n_estimators=98, max_depth=6, bagging_fraction = 0.8,feature_fraction = 0.8), 
                       param_grid = parameters, scoring='roc_auc',cv=5,n_jobs=-1)
gsearch1.fit(tr_X,tr_y) 
print('best_params:', gsearch1.best_params_, gsearch1.best_score_)
```

```python
mean_list = gsearch1.cv_results_['mean_test_score']
std_list = gsearch1.cv_results_['std_test_score']

plt.figure(figsize = (23,8),facecolor='w')

plt.subplot(1, 3, 1)
plt.plot(param, mean_list, 'ro-', markeredgecolor='k', lw=2)  
plt.ylabel('Accuracy',fontsize=12)
plt.title('Accuracy_n_estimators')
plt.grid(b=True, ls=':', color='#606060')

plt.subplot(1, 3, 2)
plt.plot(param, std_list, 'bo-', markeredgecolor='k', lw=2)  
plt.ylabel('STD',fontsize=12)
plt.title('std_n_estimators')
plt.grid(b=True, ls=':', color='#606060')
```
